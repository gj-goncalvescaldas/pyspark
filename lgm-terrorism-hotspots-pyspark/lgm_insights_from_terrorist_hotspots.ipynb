# 1. How has the number of terrorist activities changed over the years? Are there certain regions where this trend is different from the global averages?

```
attacks_per_year = df.groupBy("year") \
    .agg(F.count("*").alias("total_attacks_global")) \
    .orderBy("year")
display(attacks_per_year.limit(20))
```

```
attacks_per_year_region = df.groupBy("year", "region_txt") \
    .agg(
        F.count("*").alias("total_attacks_region"),
    ) \
    .orderBy("year", "region_txt")

joined_df = attacks_per_year_region.join(attacks_per_year, on="year", how="left")
joined_df = joined_df.withColumn(
    "region_share",
    F.round((F.col("total_attacks_region") / F.col("total_attacks_global")) * 100, 1)
)
```

```
display(joined_df.limit(20))
```

```
top_regions = df.groupBy("region_txt") \
    .agg(F.count("*").alias("total")) \
    .orderBy(F.desc("total")) \
    .limit(5)

top_regions_list = [row["region_txt"] for row in top_regions.collect()]

filtered_df = joined_df.filter(F.col("region_txt").isin(top_regions_list))

display(filtered_df.limit(20))
```

# 2. Is the number of incidents and the number of casualties correlated? Can you spot any irregularities or outliers?

```
yearl_stats = df \
  .groupBy("year") \
  .agg(
    F.count("eventid").alias("total_attacks"),
    F.sum("nkill").alias("total_killed")
  ) \
  .orderBy("year")

display(yearl_stats.limit(20))
```

```
correlation = yearl_stats.stat.corr("total_attacks", "total_killed")

print(f"Correlation between number of attacks and deaths: {correlation}")
```

```
display(yearl_stats \
    .withColumn(
        "lethality_ratio", F.col("total_killed") / F.col("total_attacks")
    ) \
    .orderBy(F.col("lethality_ratio").desc()) \
)
```

```
high_lethality_years = [1998, 2001, 2004, 2007]

df_filtered = df.filter(F.col("year").isin(high_lethality_years))


attack_types = df_filtered.groupBy("year", "attacktype1_txt").agg(
    F.count("*").alias("attack_count"),
    F.sum("nkill").cast("float").alias("total_killed"),
    F.avg("nkill").cast("float").alias("avg_killed")
).orderBy("year", F.col("total_killed").desc())

display(attack_types)
```

## ✅ Correlation Analysis

Using the yearly aggregated data:

- A **very strong positive correlation** of **0.9656** was found between the number of terrorist incidents (`total_attacks`) and the number of fatalities (`total_killed`).
- This confirms that, **in general, years with more terrorist attacks also saw more deaths**.

---

## ⚠️ Irregularities and Outliers

While the correlation is strong, an analysis of the **lethality ratio** (`total_killed / total_attacks`) highlights several years with disproportionately deadly attacks.


### 🔺 High Lethality Years
| Year | Lethality Ratio | Findings from `attacktype1_txt` |
|------|------------------|----------------------------------|
| **1998** | 5.01 | High lethality explained by a small number of extremely deadly **Bombings/Explosions**. |
| **2004** | 4.90 | Predominance of fatal **Bombings/Explosions**. |
| **2001** | 4.04 | Death toll dominated by **a few large-scale attacks**, most notably **9/11** (aviation-related suicide attacks). |
| **2007** | 3.93 | Many **Bombings/Explosions** and **Armed Assaults** with high casualty counts. |
| **2002–2006** | 3.3–3.5 | Consistently dominated by high-fatality types like **Bombings/Explosions** across multiple regions. |

These years demonstrate that **the lethality of a year depends not just on the number of attacks, but also on the type of attack carried out**.

---

## 🧠 Conclusion

- There is a strong statistical correlation between the number of terrorist attacks and total fatalities.
- However, **attack types (`attacktype1_txt`) play a critical role in explaining years with unusually high lethality**.



# 3. What are the most common methods of attacks? Does it differ in various regions or in time?

```
df_clean = df.filter(F.col("attacktype1_txt").isNotNull())


common_attacks = df_clean.groupBy("attacktype1_txt").agg(
    F.count("*").alias("total_attacks")
).orderBy(F.col("total_attacks").desc())


attacks_by_region = df_clean.groupBy("region_txt", "attacktype1_txt").agg(
    F.count("*").alias("total_attacks")
).orderBy("region_txt", F.col("total_attacks").desc())


region_window = Window.partitionBy("region_txt").orderBy(F.col("total_attacks").desc())
top_attacks_per_region = attacks_by_region.withColumn("rank", F.row_number().over(region_window)) \
    .filter(F.col("rank") <= 3)

attacks_by_year = df_clean.groupBy("year", "attacktype1_txt").agg(
    F.count("*").alias("total_attacks")
).orderBy("year", F.col("total_attacks").desc())


year_window = Window.partitionBy("year").orderBy(F.col("total_attacks").desc())
top_attacks_per_year = attacks_by_year.withColumn("rank", F.row_number().over(year_window)) \
    .filter(F.col("rank") <= 3)


print("🔹 Most common attack types overall:")
display(common_attacks)

print("🔹 Top 3 attack types per region:")
display(top_attacks_per_region.select("region_txt", "attacktype1_txt", "total_attacks"))

print("🔹 Top 3 attack types per year:")
display(top_attacks_per_year.select("year", "attacktype1_txt", "total_attacks"))
```

# 4. Plot the locations of attacks on a map to visualize their regional spread

```

df_geo = df.filter(
    (F.col("latitude").isNotNull()) & 
    (F.col("longitude").isNotNull())
).select("eventid", "year", "region_txt", "country_txt", "city", "latitude", "longitude")


df_geo_sampled = df_geo.limit(10000)

pdf_geo = df_geo_sampled.toPandas()
```

```
fig = px.scatter_geo(
    pdf_geo,
    lat="latitude",
    lon="longitude",
    color="region_txt",
    hover_name="country_txt",
    hover_data=["city", "year"],
    title="Global Terrorist Attack Locations",
    opacity=0.6,
    height=600
)

fig.update_layout(geo=dict(showland=True, showcountries=True))
fig.show()
```
